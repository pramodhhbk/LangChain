```python
#Zero-shot prompting with chain of thoughts reasoning
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.agents import Tool, initialize_agent, AgentType

# Initialize OpenAI LLM
llm = ChatOpenAI(model="gpt-4", openai_api_key="")

# Define Chains
demo_prompt = PromptTemplate(
    input_variables=["demographics"],
    template="Analyze the following demographics: {demographics}. Suggest engagement strategies."
)
demo_chain = LLMChain(llm=llm, prompt=demo_prompt)
#print(demo_chain,"demo_chain")


channel_prompt = PromptTemplate(
    input_variables=["channels"],
    template="Evaluate the following channels: {channels}. Rank and suggest usage."
)
channel_chain = LLMChain(llm=llm, prompt=channel_prompt)
#print(channel_chain,"channel_chain")

budget_prompt = PromptTemplate(
    input_variables=["budget", "channels", "demographics"],
    template="Distribute a budget of {budget} across these channels: {channels} for the demographic: {demographics}. Justify the allocation."
)
budget_chain = LLMChain(llm=llm, prompt=budget_prompt)


#print(budget_chain,"budget_chain")

# Combine Chains into Tools
tools = [
    Tool(name="DemographicAnalysis", func=demo_chain.run, description="Analyze demographics."),
    Tool(name="Channels", func=channel_chain.run, description="Optimize channels."),
    Tool(name="Budget", func=lambda x: budget_chain.run(budget=inputs['budget'], channels=inputs['channels'], demographics=inputs['demographics']), 
         description="Allocate budget. No additional input needed, uses predefined budget, channels, and demographics.")
]



# Initialize Agent
agent = initialize_agent(
    tools, llm, agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)

# Define Inputs (CASE B: Multiple Keys)
inputs = {
    'input':"What are some effective marketing strategies for Nike?",
    "demographics": "18-35, Fitness enthusiasts, Sports fans, Urban dwellers.",
    "channels": "Social Media, Influencers, Email, Events",
    "budget": "100000"
}

# Run the Agent
response = agent.run(inputs)
("Marketing Strategy Agent Response:\n", response)
```

```python
# Print Input Variables for Debugging
query1 = "What engagement strategies work best for Nike's target audience of fitness enthusiasts aged 18-35?"
response1 = agent.run({"input": query1})
print("Demographic Analysis Response:\n", response1)

```
